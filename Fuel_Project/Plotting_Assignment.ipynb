```python
plt.figure(figsize=(10, 6))
plt.hist(data['co2'], bins=10, color='green', edgecolor='black')
plt.title('Histogram of CO2 Emissions (10 bins)')
plt.xlabel('CO2 Emissions (g/mile)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['co2'], bins=20, color='blue', edgecolor='black')
plt.title('Histogram of CO2 Emissions (20 bins)')
plt.xlabel('CO2 Emissions (g/mile)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['co2'], bins=30, color='purple', edgecolor='black')
plt.title('Histogram of CO2 Emissions (30 bins)')
plt.xlabel('CO2 Emissions (g/mile)')
plt.ylabel('Frequency')
plt.show()
```

```python
print(data.dtypes)

data_numeric = data.select_dtypes(include=[np.number])

print(data_numeric.isnull().sum())

data_numeric = data_numeric.dropna()

correlation_matrix = data_numeric.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Correlations Between Numeric Variables')
plt.savefig('histogram_fuel_efficiency.png')
plt.savefig('correlation_heatmap.png')
plt.show()
```
